```python
from networkx.generators.lattice import grid_graph

#networkx.generators.lattice.grid_graph
G = grid_graph(dim=[2, 3, 4])
len(G)

G = grid_graph(dim=[range(0, 5), range(0, 5)])
len(G)
```

```python
import networkx as net
import matplotlib.pyplot as plot

net.draw(G,  with_labels = True, node_color='white', edge_color = 'blue', width = 1, node_size= 600)
plot.show()
```
